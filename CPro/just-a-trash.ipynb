```python
import pandas as pd
```

```python
df = pd.read_csv('DelayedFlights.csv')
df
```

```python
df = df.drop(columns=['Unnamed: 0', 'Year', 'Month', 'UniqueCarrier'])
```

```python
print(df[df['Cancelled'] != 0].shape[0])
print(df[df['CancellationCode'] != "N"].shape)
```

```python
df[df['Diverted'] != 0].shape[0]
```

```python
df[['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']].describe()
```

```python
df.isna().sum()
```

```python
df[['TailNum','CRSElapsedTime','TaxiOut','ArrTime', 'ActualElapsedTime', 'AirTime', 'ArrDelay', 'TaxiIn']] = (
    df[['TailNum','CRSElapsedTime','TaxiOut','ArrTime', 'ActualElapsedTime', 'AirTime', 'ArrDelay', 'TaxiIn']].fillna(0)
)
df.isna().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cmap='viridis')
plt.title('Non a number datas')
```

```python
(df[['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']] != 0).sum()
```

```python
delay_cols = ['CarrierDelay', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay']
df[delay_cols] = df[delay_cols].fillna(0)
```

```python
df[delay_cols].isna().sum()
```

```python
total_delays = df[delay_cols].sum().sort_values(ascending=False)
print(total_delays)
```

```python
plt.figure(figsize=(8,5))
plt.bar(total_delays.index, total_delays.values)
plt.title('Total Delay Minutes by Cause')
plt.ylabel('Total Delay (minutes)')
plt.xlabel('Delay Cause')
plt.xticks(rotation=30)
plt.grid(axis='y', alpha=0.7)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Distance'], df['ArrDelay'], alpha=0.4)
plt.title('Distance vs Arrival Delay')
plt.xlabel('Distance (miles)')
plt.ylabel('Arrival Delay (minutes)')
plt.grid(True, linestyle='--', alpha=0.6)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Distance'], df['ArrDelay'], alpha=0.4)
plt.title('Distance vs Arrival Delay')
plt.xlabel('Distance (miles)')
plt.ylabel('Arrival Delay (minutes)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.xlim(0, 3000)
plt.ylim(-150, 800)
```

```python
print(df[['Distance', 'ArrDelay']].corr())
print('\n\nFlight distance doesn\'t strongly influence how delayed a flight is')
```

```python
corr = df.corr(numeric_only=True)
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".1f", linewidths=0.5)
plt.title("Correlation Heatmap of the Dataset")
```

```python
colors = df['ArrDelay'].apply(lambda x: 'green' if x < 0 else 'red')

plt.figure(figsize=(8,5))
plt.scatter(df['Distance'], df['ArrDelay'], c=colors, alpha=0.5)
plt.title('Distance vs Arrival Delay (Early vs Delayed Flights)')
plt.xlabel('Distance (miles)')
plt.ylabel('Arrival Delay (minutes)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.scatter([], [], color='green', label='Early Arrival')
plt.scatter([], [], color='red', label='Delayed Arrival')
plt.legend()
```
